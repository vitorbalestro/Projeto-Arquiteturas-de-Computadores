Models

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
```

Imports

```python
from sklearn.preprocessing import LabelEncoder
from joblib import dump
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.pipeline import make_pipeline
```

Preprossesing items

```python
tweet_data = pd.read_csv('open_twitter_human_bots_dataset.csv',index_col=0)
sc = StandardScaler()
enc = LabelEncoder()
id_list = tweet_data['id']
descriptions = tweet_data['description']
names = tweet_data['screen_name']
types = tweet_data['account_type']
tweet_data
```

Preprossesing

```python
is_Description_Null = []
Description_Lenght = []
for i in range(len(descriptions)):
    Description_Lenght.append(len((str)(descriptions[i])))
    is_Description_Null.append(len((str)(descriptions[i])) > 3)

is_bot = []
for i in range(len(types)):
    is_bot.append((str)(types[i]) == 'bot')

Name_Lenght = []
for i in range(len(names)):
    Name_Lenght.append(len(names[i]))

tweet_data["is_Description_Null"] = is_Description_Null
tweet_data["Description_Lenght"] = Description_Lenght
tweet_data["Name_Lenght"] = Name_Lenght
tweet_data["is_bot"] = is_bot
```

Start Training

```python
X = tweet_data.drop(columns=['split','profile_background_image_path','profile_image_path','account_type','profile_background_image_url','profile_image_url','id','screen_name','description','lang','location','created_at','is_bot'])
y = tweet_data['is_bot']
clf = make_pipeline(StandardScaler(),SVC(verbose=True,probability=True,break_ties=True))
X_train,X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
clf.fit(X_train, y_train)
pref = clf.predict(X_test)
accuracy_score(y_test,pref)
```

Serializing AI

```python
dump(clf, 'SVC_tweet.pkl')
```

```python
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
predictions = model.predict(X_test)
accuracy_score(y_test,predictions)
```
